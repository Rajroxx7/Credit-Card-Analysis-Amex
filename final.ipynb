```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("development_dataset.csv")
```

```python
final=pd.read_csv("Evaluation_dataset.csv")
```

```python
final.head()
```

```python
train.head()
```

```python
data5=train
del data5['VAR21']
data5=[final,data5]
comb1=pd.concat(data5)
```

```python
comb1['VAR14']=comb1['VAR14'].replace('.','1')
comb1['VAR2'].fillna(comb1['VAR2'].mean(),inplace=True)
#data[data['VAR3']<0.1].VAR3.hist()
comb1['VAR5'].fillna(comb1['VAR5'].value_counts().idxmax(),inplace=True)
comb2=comb1.drop('VAR17',axis=1)
comb2=comb2.drop('VAR9',axis=1)
comb2['VAR6'].fillna(comb2['VAR6'].median(),inplace=True)
comb2['VAR7'].fillna(comb2['VAR7'].median(),inplace=True)
comb2['VAR8'].fillna(comb2['VAR8'].median(),inplace=True)
comb2['VAR10'].fillna(comb2['VAR10'].mean(),inplace=True)
comb2['VAR11'].fillna(comb2['VAR11'].median(),inplace=True)
comb2['VAR12'].fillna(comb2['VAR12'].mean(),inplace=True)
comb2['VAR13'].fillna(comb2[comb2['VAR13']<20].VAR13.mean(),inplace=True)
comb2['VAR15'].fillna(comb2[comb2['VAR15']<40].VAR15.mean(),inplace=True)
comb2['VAR16'].fillna(comb2.VAR16.median(),inplace=True)
comb2['VAR4'].fillna(comb2.VAR4.median(),inplace=True)
t=comb2['VAR3']
t=np.log(t)
t.fillna(t.mean(),inplace=True)
z=np.exp(t)
comb2['VAR3']=z
comb2.isnull().sum()
```

```python
test2=comb2.iloc[0:10000,:]
test2.head()
```

```python
train2=comb2.iloc[10000:44000,:]
train2.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=train2.drop('VAR1',axis=1)
```

```python
train99=pd.read_csv("development_dataset.csv")
```

```python
y=train99['VAR21']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=69)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
import xgboost as xgb
```

```python
model=xgb.XGBClassifier()
```

```python
model.fit(X_train.drop('VAR14',axis=1),y_train)
```

```python
pred=model.predict(X_test.drop('VAR14',axis=1))
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(pred,y_test)
```
